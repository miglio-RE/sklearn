```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder
from itertools import product
```

```python
data=pd.read_csv("./mushrooms.csv", na_values=["?"])
data
```

```python
data.info()
```

```python
colonne=data.columns
colonne
```

```python
colonne_numeriche={}
le_dict={}
for col in colonne:
    #trasforma colonne qualitatitive in colonne numeriche
    le_dict[col] = LabelEncoder()
    colonne_numeriche[col]=le_dict[col].fit_transform(data[col])
colonne_numeriche
nuovo_data=pd.DataFrame(colonne_numeriche)
nuovo_data
le_dict
```

```python
le_dict["habitat"].classes_
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X=nuovo_data[colonne[:-1]].values
y=nuovo_data["class"].values
X_train, X_test, y_train, y_test=train_test_split(X, y, train_size=0.7, random_state=0)
# Creazione di un oggetto StandardScaler
scaler = StandardScaler()

# Adattamento e trasformazione dei dati
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
nuovo_data.columns[1:]
```

```python
modello_finale=None
diz={}
massimo=0
while massimo!=1.0:  
    model=MLPClassifier(hidden_layer_sizes=(10, 10), random_state=1, max_iter=50)
    colonne_test=np.random.choice(colonne[1:], 5, True)
    
    
    X=nuovo_data[colonne_test].values
    y=nuovo_data["class"].values
    X_train, X_test, y_train, y_test=train_test_split(X, y, train_size=0.7, random_state=0)
    modello_finale=model.fit(X_train, y_train)
    if model.score(X_test, y_test)>0.99:
        massimo=model.score(X_test, y_test)
        diz[str(colonne_test)]=massimo
        
       
        print(colonne_test)
        print(f"Accuratezza set TRAIN: {model.score(X_train, y_train)}")
        print(f"Accuratezza set TEST: {model.score(X_test, y_test)}")
    
diz
   
```
