```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("./fer2013.csv")
nuovo_data=data[((data["emotion"]==0)|(data["emotion"]==2))&((data["Usage"]=="PublicTest")|(data["Usage"]=="Training"))][:1000].copy()
nuovo_data["Usage"].unique()
```

```python
file=open("./fer2013.csv")
righe=file.readlines()
file.close()
righe=righe[1:1000]
emozioni=[]
facce=[]
for riga in righe:
    riga=riga.split(",")
    emozioni.append(int(riga[0]))
    immagine=riga[1].split(" ")
    facce.append(np.array(immagine))
                    
```

```python
array_facce=np.array(facce)
data_facce=pd.DataFrame()
data_facce["emozione"]=emozioni
for j, faccia in enumerate(facce):
    for i in range(len(faccia)):
        if (emozioni[j]==0) or (emozioni[j]==1):
            data_facce[f"pixel_{i}"]=faccia[i]
data_facce
        
```

```python
data_facce[(data_facce["emozione"]==0)|(data_facce["emozione"]==2)]
```

```python
colonne=data_facce.columns
colonne[1:]
```

```python
X=data_facce[colonne[:-1]].values
y=data_facce["emozione"].values
X_train, X_test, y_train, y_test=train_test_split(X, y, train_size=0.7, random_state=0)
model=MLPClassifier(hidden_layer_sizes=(1500, 1500, 1500), random_state=1, max_iter=300)
model.fit(X_train, y_train)
print(f"Accuratezza set TRAIN: {model.score(X_train, y_train)}")
print(f"Accuratezza set TEST: {model.score(X_test, y_test)}")
       
```
