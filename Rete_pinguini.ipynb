```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("./penguins_size.csv", na_values=".")
data
```

```python
data.info()
```

```python
import seaborn as sns#si basa su matplotlib
#analisi grafica
sns.set_theme(font_scale=2)
sns.pairplot(data, hue="species", dropna=True)
sns.set()
```

```python
data_puliti=data.dropna()
data_puliti.info()
```

```python
diz_sesso={}
for i, sesso in enumerate(data_puliti["sex"].unique()):
    diz_sesso[sesso]=i
diz_sesso
    
```

```python
diz_isola={}
for i, specie in enumerate(data_puliti["island"].unique()):
    diz_isola[specie]=i
diz_isola
```

```python
lista_sesso=[]
for dato in data_puliti.values:
    lista_sesso.append(diz_sesso[dato[6]])
```

```python
lista_isola=[]
for dato in data_puliti.values:
    lista_isola.append(diz_isola[dato[1]])
```

```python
data_puliti["sesso"]=np.array(lista_sesso)

data_puliti["mumero_isola"]=np.array(lista_isola)
data_puliti
```

```python
selected_cols=["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]#colonne di interesse
d=data.dropna(axis="index", subset=selected_cols)
data.info()
```

```python
import seaborn as sns#si basa su matplotlib
#analisi grafica
sns.set_theme(font_scale=2)
sns.pairplot(data_puliti, hue="species", dropna=True)
sns.set()
```

```python
X=data_puliti[["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]].values
y=data_puliti["species"].values
X_train, X_test, y_train, y_test=train_test_split(X, y, train_size=0.7, random_state=0)
# Creazione di un oggetto StandardScaler
scaler = StandardScaler()

# Adattamento e trasformazione dei dati
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
fig, axes = plt.subplots(3, 2, figsize=(10, 17), sharey="row")

for i, ax_row in enumerate(axes):
    col = selected_cols[i]
    print(col)
    ax_row[0].set_xlabel("dati iniziali")
    ax_row[1].set_xlabel("dati normalizzati")
    ax_row[0].set_ylabel(f"{col} - numero di campioni")
    sns.histplot(ax=ax_row[0], x=X_train_scaled[:, i], kde=True)
    sns.histplot(ax=ax_row[1], x=X_test_scaled[:, i], kde=True)

sns.set()
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
model=MLPClassifier(hidden_layer_sizes=(10, 10), random_state=1, max_iter=300)
#hidden_size: numero di nodi nascosti 100 100
#max iter. diciamo al modello di allenarsi 300 volte. impara sui set di allenamento
#per 300 volte scorriamo l'array e per ogni riga specifichiamo le carattteristiche dell'alieno
```

```python
model.fit(X_train_scaled, y_train)
```

```python
print(f"Accuratezza set TRAIN: {model.score(X_train_scaled, y_train)}")
print(f"Accuratezza set TEST: {model.score(X_test_scaled, y_test)}")
```
